```python
# Q. Write a program to print sentence/word is palindrom or not.
'''
1. Ask the user to enter the string.
2. store the string in variable: string
3. reverse the  entered string : string1
4. use is else condition to check the string is palindrome or not.

'''
# Step1 : Ask the user to enter the string.
string = input("Enter the String : ")

# Step2 : reverse the  entered string : string1
string1 = string[::-1]

# Step3 : use if else logic
if string == string1:
  print("The Given String is Palindrome")
else:
  print("The Given String is not Palindrome")
```

```python
# Number Guessing Game: With Latency Calculation (elif)
import time
import random

start = time.perf_counter()

number = random.randint(1,5)
print(number)

if number == 1:
  print("Number is 1")
elif number == 2:
  print("Number is 2")
elif number == 3:
  print("Number is 3")
elif number == 4:
  print("Number is 4")
elif number == 5:
  print("Number is 5")
else:
  print("No Number")

# Your code that you want to time
end = time.perf_counter() - start
print('{:.6f}s for the calculation'.format(end))

```

```python
# Calculate length of each string

# define a iterable: List
lst = ["Hello","World","PythonProgramming","Datascience","Java"]

def compute1():      # Function Defination
  bucket1 = []
  bucket2 = []

  for i in lst:
    if len(i) > 4:
      bucket1.append(i)
    else:
      bucket2.append(i)
  return bucket1,bucket2

bucket1,bucket2 = compute1()         # Function Calling
print(bucket1)
print(bucket2)
```

```python
lst1 = ["a","e",'h',"o","u"]

bucket1 = []
bucket2 = []

def compute2():
  for i in lst1:
    if i in "aeiou":
      bucket1.append(i)
    else:
      bucket2.append(i)
  return bucket1,bucket2

bucket1,bucket2 = compute2()
print(bucket1)
print(bucket2)
```
